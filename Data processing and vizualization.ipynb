```python
#2.1

df = pd.read_csv(r'C:\Users\admin\Desktop\Настя\shopping_trends.csv', encoding="latin1")
```

```python
# Перевірити розмір таблиці
table_size = df.shape

# Перевірити наявність пропусків
missing_values = df.isnull().sum()

# Перевірити типи стовпців
column_types = df.dtypes

# Перейменувати стовпці
df.columns = df.columns.str.lower().str.replace(' ', '_')

print(f"Розмір таблиці: {table_size}")
print(f"Пропуски в даних:\n{missing_values}")
print(f"Типи стовпців:\n{column_types}")
print(f"Перейменовані стовпці:\n{df.columns}")
```

```python
import seaborn as sns
# Побудувати гістограму віку
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', bins=30, kde=True, color='skyblue')
plt.title('Гістограма віку')
plt.xlabel('Вік')
plt.ylabel('Частота')
plt.show()
```

```python
#2.2

# Побудувати графік порівняння розподілу покупок за статтею
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='item_purchased', order=df['item_purchased'].value_counts().index, palette='viridis')
plt.title('Порівняння розподілу покупок за статтею')
plt.xlabel('Стаття товару')
plt.ylabel('Кількість покупок')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#2.3

# Побудувати графік розподілу всіх товарів
plt.figure(figsize=(15, 8))
sns.countplot(data=df, x='item_purchased', order=df['item_purchased'].value_counts().index, palette='viridis')
plt.title('Розподіл покупок за кожним товаром')
plt.xlabel('Стаття товару')
plt.ylabel('Кількість покупок')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#2.4

# Побудувати scatter plot для залежності між віком та сумою покупок
plt.figure(figsize=(12, 8))
sns.barplot(data=df, x='age', y='purchase_amount_(usd)', hue='gender', palette='viridis', alpha=0.7)
plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупок')
plt.show()
```

```python
#2.5

plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='age', y='purchase_amount_(usd)', hue='gender', palette='viridis', alpha=0.7)
plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупок')
plt.show()
```

```python
#2.6

# Побудувати графік розподілу оцінок товарів
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='review_rating', palette='viridis')
plt.title('Графік розподілу оцінок товарів')
plt.xlabel('Оцінка товару')
plt.ylabel('Кількість оцінок')
plt.show()
```

```python
#2.7

# Побудувати графік розподілу оцінок якості товарів
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='review_rating', bins=10, kde=True, color='skyblue')
plt.title('Графік розподілу оцінок якості товарів')
plt.xlabel('Оцінка якості товару')
plt.ylabel('Кількість оцінок')
plt.show()
```

```python
#2.8

# Побудувати категоріальну діаграму для статі
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='gender', palette='viridis')
plt.title('Категоріальна діаграма для статі')
plt.xlabel('Стать')
plt.ylabel('Кількість')
plt.show()
```

```python
#2.10

# Побудувати scatter plot для залежності між рейтингом і сумою покупок
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='review_rating', y='purchase_amount_(usd)', hue='gender', palette='viridis', alpha=0.7)
plt.title('Залежність між рейтингом і сумою покупок')
plt.xlabel('Рейтинг')
plt.ylabel('Сума покупок')
plt.show()
```
